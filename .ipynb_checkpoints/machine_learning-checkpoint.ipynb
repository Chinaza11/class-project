```python
np.random.seed(8000)
con = sqlite3.connect('mydb.db')
df = pd.read_sql('select * from treatments',con)
display(df)
con.close()
```

```python
df = df.drop(columns=['treatments_id', 'year', 'replicate', 'latitude', 'longitude'])
display(df)
df.info() #to check for null values, none found
```

```python
bins = [0,50,200]
labels = ['low','high']
```

```python
df['fruit_cat'] = pd.cut(df['fruit_yield'],bins,labels = labels)
df
```

```python
df['fruit_cat'].value_counts()
```

```python
df = df.drop(columns=['fruit_yield'])
df
```

```python
X = df.drop(columns=['fruit_cat'])
y = df['fruit_cat']
```

```python
X_train,X_test,y_train,y_test = model_selection.train_test_split(X,y,train_size=.7,stratify=df['fruit_cat'])

#Concatenating the feature and target data into a dataframe and visualizing them to check for weird abnormalities like outliers
df_train = pd.concat([X_train,y_train],axis=1)
#sns.pairplot(df_train)
```

```python
enc = preprocessing.OneHotEncoder(sparse=False,drop='if_binary')
fit_enc = enc.fit(X_train[['treatments_name']])
fit_enc = fit_enc.transform(X_train[['treatments_name']])
```

```python
X_train_dummies = pd.get_dummies(X_train,columns=['treatments_name'])
X_train_dummies
```

```python
rf = ensemble.RandomForestClassifier(random_state=8000)
gscv_rf = model_selection.GridSearchCV(rf, param_grid={"n_estimators":range(2,100)},cv=10,verbose=10,n_jobs=-1)
gscv_rf.fit(X_train_dummies,y_train)
```

```python
gscv_rf.score(X_train_dummies,y_train)
```

```python
X_test_dummies = pd.get_dummies(X_test, columns= ['treatments_name'])
X_test_dummies
```

```python
y_test_pred = gscv_rf.predict(X_test_dummies)
print(metrics.classification_report(y_test,y_test_pred))
```

```python
print(metrics.confusion_matrix(y_test,y_test_pred))
```
